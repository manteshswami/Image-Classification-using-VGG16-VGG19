```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Downloading & Loading the VGG16 Model
model_1= VGG16(weights='imagenet')
print(model_1.summary())
```

```python
from tensorflow.keras.applications import VGG19
model_2=VGG19(weights='imagenet')
print(model_2.summary())
```

### Prediction Using pretrained Model

```python
def predict_img(model,img_path):

    ## load the img with required target size
    img=load_img(img_path,target_size=(224,224))

    ## convert img to numpy array
    img_array=img_to_array(img)

    ## add a batch dimension (1,224,224,3)
    img_array=np.expand_dims(img_array,axis=0)

    ## preprocess the i/p
    img_array=preprocess_input(img_array)

    ## predict the img class
    prediction=model.predict(img_array)

    ## decoding the prediction
    label=decode_predictions(prediction,top=5)
    return label[0]


```

### Passing img for prediction

```python
sample_img_path='/content/green_mamba.jpg'

## Prediction the img class
predictions=predict_img(model_1,sample_img_path)

## disply the prediction
for i, (imagenet_id,label,score) in enumerate(predictions):
    print(f"{i+1}. {label} ({score*100:.2f}%)")
```

```python
img=load_img(sample_img_path)
```

```python
plt.imshow(img)
plt.title(f"Label={predictions[0][1]} and Score={predictions[0][2]:.2f}%")
plt.axis('off')
plt.show()
```
